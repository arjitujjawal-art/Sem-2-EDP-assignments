```python
# Tokenization

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

text = "Tokenization groups the words into tokens and stores them"

tokens = word_tokenize(text)

print(tokens)
```

```python
# Sentence Tokenization

from nltk.tokenize import sent_tokenize

text = "Symbiosis Institute of Technology is one of the top private college in Pune. It has a good ranking."

# with or without full stop at the end of last sentence it doesn't make any difference

sentences = sent_tokenize(text)

print(sentences)
```

```python
# Stop Word Removal
# Stop words such as is, the, and, in are removed

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

words = word_tokenize(text)

filtered_words = [w for w in words if w.lower() not in stop_words]

print(filtered_words)
```

```python
# Stemming
# Stemming reduces words to their root form

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["Athletics", "Basketball", "Kho - Kho", "Kabaddi", "Throwball"]

for w in words:
  print(stemmer.stem(w))
```

```python
# Lemmatization
# Lemmatization converts words to their Dictionary form

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("was"))
print(lemmatizer.lemmatize("plays"))
```

```python
# Part of Speech (POS) Tagging
# POS tagging identifies grammatical roles of words.

import nltk
nltk.download('average_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag


words = word_tokenize("This is an example sentence which is used in Python Programming")

pos_tags = pos_tag(words)

print(pos_tags)
```

```python
# Word Frequency Count

from nltk.probability import FreqDist

words = word_tokenize(text)

fd = FreqDist(words)

print(fd.most_common())
```
