```python
# To Create a Dataset
import pandas as pd
data={
    'Order ID':['01','02','03','04','05','06','07','08','09','10'],
    'Category':['Furniture','Office Supplies','Furniture','Technology','Technology','Furniture','Office Supplies','Technology','Furniture','Technology'],
    'Payment Method':['Cash','Credit Card','Cash','Cash','Credit Card','Cash','Credit Card','Cash','Cash','Credit Card'],
    'Delivery Type':['Standard Class','First Class','Second Class','Express','Second Class','Standard Class','Express','First Class','Standard Class','Standard Class'],
    'Customer Type':['New','Returning','New','Returning','New','Returning','New','Returning','New','Returning']
}
df=pd.DataFrame(data)
df
```

```python
# Frequency Count
# Count how many orders in each category.
print(df['Category'].value_counts())
```

```python
# Frequency count of Payment Method
print(df['Payment Method'].value_counts())
```

```python
# Unique values
# Display all unique categories.
print(df['Category'].unique())
```

```python
# Number of Unique Values (nunique)
print(df['Category'].nunique())
```

```python
# Number of Unique Values (nunique) for Payment Method
print(df['Payment Method'].nunique())
```

```python
# Cross Tabulation
# Category vs Payment Method
print(pd.crosstab(df['Category'],df['Payment Method']))
```

```python
# Cross Tabulation (Delivery vs Customer Type)
print(pd.crosstab(df['Delivery Type'],df['Customer Type']))
```

```python
# Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
# Filtering Data
# Example: Orders from Furniture Cateogry
furniture_orders=df[df['Category']=='Furniture']
print(furniture_orders)
```

```python
# Technology Category
technology_orders=df[df['Category']=='Technology']
print(technology_orders)
```

```python
# Grouping
#Orders by category and payment method
print(df.groupby('Category')['Payment Method'].value_counts())
```

```python
# Sorting Data(Sorted Alphbetically)
sort_df=df.sort_values(by='Category')
print(sort_df)
```
