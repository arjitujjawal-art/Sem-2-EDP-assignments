```python
# Dataset Creation

import pandas as pd

data = {
    "Roll_No": [101, 102, 103, 104, 105],
    "Gender": ["Female", "Male", "Female", "Male", "Female"],
    "Department": ["CSE", "ECE", "MECH", "CIVIL", "ENTC"],
    "CGPA" : [7.98, 9.00, 7.01, 10.00, 8.3],
}

df = pd.DataFrame(data)

df.to_csv("student_data.csv", index=False)

df
```
