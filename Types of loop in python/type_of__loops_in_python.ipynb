###Types of Loops in  Python


```python
#while loop
#used when the number of itertions is not known
#Print i as long as i is less than 6:
i=0
while i<6:
  print(i)
  i+=1
```

```python
#print number from 1 to N
inp=int(input("Enter the value of N:"))
i=1
while i<=inp:
  print(i)
  i+=1
```

```python
#factorial of a number:
inp=int(input("Enter the value of N:"))
k,i=inp,1
while k>0:
  i=i*k
  k=k-1
print(f"factorial of {inp} is",i)
```

```python
#fibonacci series while loop
n=int(input("Enter the value of N:"))
a,b,i=0,1,0
while i<n:
  print(a,end=',')
  c=a+b
  a=b
  b=c
  i+=1
```

```python
#reverse a number
num=int(input("Enter the number:"))
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num//=10
print("reverse number:",rev)
```
