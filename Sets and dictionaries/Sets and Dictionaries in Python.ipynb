```python
#creation of dict
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
print(thisdict)
```

```python
#print key values
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
print(thisdict["colour"])
```

```python
#Dict will not allow duplicate
thisdict={"food":"pizza","colour":"red","name":"ajrit","name":"rohan"}
print(thisdict)
```

```python
#length of the dictionary
len(thisdict)
```

```python
#Changing/adding key value in dict
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
thisdict["colour"]="blue"
print(thisdict)
thisdict["number"]="45"
print(thisdict)
```

```python
#retrival
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
print("food:",thisdict["food"])
```

```python
#adding and updating
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
thisdict["colour"]="blue"
print(thisdict)
thisdict["number"]="45"
print(thisdict)
```

```python
#removal
thisdict={"food":"pizza","colour":"red","name":"ajrit"}
print(thisdict)
thisdict.pop("food")
print("after removal",thisdict)
```
