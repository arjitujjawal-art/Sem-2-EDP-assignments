```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [50, 70, 30, 90, 60, 55],
    'Sales' : np.random.randint(100,500,6),
    'Profit' : np.random.randint(10,100,6)
})

df
```

```python
# 1. Area Plot

plt.figure()
plt.fill_between(df['Category'], df['Values'], color = 'blue', alpha = 0.25)
# Alpha changes the intensity of the colour

plt.title("Area Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# 2. Pie Chart

plt.figure()
plt.pie(df['Values'], labels = df['Category'], autopct = '%1.4f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# Broken Pie Chart

plt.figure()
explode = (0, 0.1, 0, 0,0.1, 0)
plt.pie(df['Values'],explode = explode, labels = df['Category'], autopct = '%1.4f%%', startangle = 90)
plt.title("Pie Chart")
plt.show()
```

```python
# 3. Donut Chart

plt.figure()
plt.pie(df['Values'], labels = df['Category'], autopct = '%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc = 'white') # here 0.7 is the radius of the inner circle
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

```python
# 4. Boxplot (outliers detection)

plt.figure()
sns.boxplot(x = df['Values'])

# if the data has outliers then the graph will move from the centre

plt.title("Boxplot of Sales")
plt.show()
```

```python
# 5. Heatmap (Correlation)

plt.figure()
corr  = df[['Sales', 'Profit']].corr()

sns.heatmap(corr, annot = True, cmap = 'plasma')

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Bubble Plot

plt.scatter(df['Sales'], df['Profit'], s = df['Values'] * 150, color = 'darkorange', edgecolors= 'black', linewidths= 2) # Size of bubbles

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize = (6,4))

sns.scatterplot(x = 'Sales', y = 'Profit', size = 'Values', hue = 'Values',
                data = df, sizes = (50,300), palette = 'viridis')

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")
```

```python
# 2.Second Dataset

np.random.seed(7)
n = 100

df = pd.DataFrame({
    'Customer_ID' : range(1, n+1),
    'Age' : np.random.randint(20,60,n),
    'Income' : np.random.randint(20000,100000,n),
    'Loan_Amount' : np.random.randint(5000, 50000, n),
    'Credit_Score' : np.random.randint(300,850,n)
})

df
```

```python
# 1. Area Plot

plt.subplot(3, 2, 1)
df_sorted = df.sort_values('Age').reset_index()
plt.fill_between(df_sorted.index, df_sorted['Income'], label='Income', alpha=0.5, color='skyblue')
plt.fill_between(df_sorted.index, df_sorted['Loan_Amount'], label='Loan Amount', alpha=0.5, color='orange')
plt.title(' Area Plot: Income vs Loan Amount')
plt.xlabel('Index (Sorted by Age)')
plt.ylabel('Amount ($)')
plt.legend()
```

```python
# 2. Pie Chart

plt.figure()
plt.pie(df['Income'], labels = df['Credit_Score'], radius = 4)
plt.title("Pie Chart")
plt.show()
```

```python
# 4. Heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'Income', 'Loan_Amount', 'Credit_Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
# 5. Bubble Plot

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Income'], s=df['Loan_Amount']/100, alpha=0.5, c=df['Credit_Score'], cmap='viridis')
plt.colorbar(label='Credit Score')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income (Size=Loan, Color=Credit Score)')
```
