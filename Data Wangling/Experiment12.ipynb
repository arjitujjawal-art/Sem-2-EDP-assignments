```python

import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Arjit", "Meera", "Karan", "Bedang"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
print(df)
df.to_csv(("student_data.csv"),index=False)
```

```python
# Step 2: Replace '-' with NaN to mark missing values
import numpy as np

df.replace("-", np.nan, inplace=True)

# Convert columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')

'''When converting data (like strings -> datetime),
if some values are invalid or cannot be converted,
instead of throwing an error, it will:
Replace those invalid values with NaT (Not a Time)'''
df
```

```python
#Step3:check missing values

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# fill missing age with mean
df["Age"].fillna((df["Age"]).mean(),inplace=True)
df
```

```python
# Fill marks missing with median
df["Marks"].fillna((df["Marks"]).median(),inplace=True)
df
```

```python
#fix omcpmsostemt text formating
df["Department"] = df["Department"].str.upper()
df
```

```python
#covert admiisio date to standard date time format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"],errors="coerce")
#'''if valur cannot be converted to a valid date , pandas will replace it with NaT(Not a Time) in'''
df
```

```python
#save cleaned data
df.to_csv("cleaned_data_.csv", index=False)
df
```
